Grupowanie przydaje się do:
    
    Podziału danych na podgrupy
    
    Stosowania innych funkcji do każdej grupy
    
    Łączenia danych w inne strukutry

```python
pogrupowane = z_csv.groupby("Year")
pogrupowane
```

```python
for group in pogrupowane:
    print(group)
```

<h3> Serie czasowe </h3>

Pandas posiada szereg narzędzi do wygodnej pracy z seriami czasowymi - w tym funkcje kroczące (powyżej) oraz agregacji i grupowania.

```python
#Stwórzmy fikcyjny zbiór danych
seria = pd.date_range('1/06/2019', periods=168, freq="H")
ts = pd.Series(np.random.randint(0,500,168), index=seria)
ts
```

```python
pd.to_datetime([1349720105, 1349806505, 1349892905, 1349979305, 1350065705], unit='ms') #epoki też działają
```

<h3> Wstęp do wykresów - wygodny pandas </h3>

```python
ran = [np.random.randint(1,200) for n in range(168)]
line = [n*2 for n in range(168)]
poly_ran = [np.random.randint(1,100) for n in range(168)]
a = [np.random.randint(1,300) for n in range(168)]
```

```python
stacked = np.vstack((ran,line,poly_ran,a)).T
df = pd.DataFrame(stacked, index = ts.index, columns = [1,2,3,4])
df = df.cumsum()
plt.figure(figsize=(14,4))
plt.plot(df)
```
